```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import math
import matplotlib 
# перевод из полярных координат в декартовые
def polar_dec(r, fi):
    x = r*np.cos(fi)
    y = r*np.sin(fi)
    return x,y
print(polar_dec (15, np.pi/4))

# построение круга и отрезка
x = np.linspace(0, 2*np.pi, 100)
r = np.linspace(50,50,100)
plt.polar(x, r)

x = np.linspace(np.pi/4, np.pi/4, 2)
r = np.linspace(10,100,2)
plt.polar(x, r)
```
