```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import math
import matplotlib 
import pylab
# построение окружности
def draw_circle(r=2, x0=0, y0=0, color='blue'):
    angle = np.linspace(0, 2*np.pi, 360)
    
    x = np.add(r*np.cos(angle), x0)
    y = np.add(r*np.sin(angle), y0)

    plt.plot(x, y, color=color, linewidth=1)
    plt.gca().set_aspect('equal')
    plt.xlabel("x")
    plt.ylabel("y")
draw_circle(x0=0, y0=0)
draw_circle(x0=2, y0=2)
```

```python
#построение эллипса
def draw_ellips(r1=1, r2=2, x0=0, y0=0, color='green'):
    angle = np.linspace(0, 2*np.pi, 360)
    
    x = np.add(r1*np.cos(angle), x0)
    y = np.add(r2*np.sin(angle), y0)
    
    plt.plot(x, y, color=color, linewidth=1)
    plt.gca().set_aspect('equal')
    plt.xlabel("x")
    plt.ylabel("y")
draw_ellips(x0=0, y0=0)
```

```python
# построение гиперболы
def draw_hyper(a, b, x0=0, y0=0, color='green'):
    x1 = []
    x2 = []
    y = []
    for t in np.linspace(-2, 2, 100):
        x1.append(a * math.cosh(t) + x0)
        x2.append(-a * math.cosh(t) + x0)
        y.append(b * math.sinh(t) + y0)
    plt.plot(x1, y)
    plt.plot(x2, y)
   
    plt.show()

draw_hyper(2, 3, 1, 1)
```
