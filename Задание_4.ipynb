```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import math
import matplotlib 
import pylab
from scipy.optimize import fsolve

def sin_ax(a,x):
    return np.sin(a*x)

a = 0.019
x = np.linspace(100, 500, 1000)
y = sin_ax(a,x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()

x0 = float(fsolve(sin_ax,200,a))
x1 = float(fsolve(sin_ax,300,a))
x2 = float(fsolve(sin_ax,500,a))
x = [x0*(x0<500)*(x0>100),x1*(x1<500)*(x1>100),x2*(x2<500)*(x2>100)]
if 0 in x:
    x.remove(0)
print('Корни уравнения: ',list(set(x)))
```
