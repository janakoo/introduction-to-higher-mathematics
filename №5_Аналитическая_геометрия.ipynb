```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import math
import matplotlib 
# параллельные плоскости
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid (X, Y)
Z1 = 2*X + 0.5*Y
Z2 = 2*X + 0.5*Y + 100
ax.plot_wireframe(X,Y,Z1, color='grey')
ax.plot_wireframe(X,Y,Z2, color='blue')
ax.scatter(0, 0, 0, 'z', 150, 'red')
plt.xlabel('x')
plt.ylabel('y')
show()
```

```python
# посторение поверхностей 
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid (X, Y)
Z = 2*X**2 + 5*Y+100
ax.plot_surface(X,Y,Z, color='pink')
ax.scatter(0, 0, 0, 'z', 150, 'black')
plt.xlabel('x')
plt.ylabel('y')
show()
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid (X, Y)
Z = X**3 + 5*Y**2-50
ax.plot_surface(X,Y,Z, color='green')
ax.scatter(0, 0, 0, 'z', 150, 'red')
plt.xlabel('x')
plt.ylabel('y')
show()
```
