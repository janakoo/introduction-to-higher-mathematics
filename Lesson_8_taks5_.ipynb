```python
import numpy as np 
import matplotlib.pyplot as plt
import scipy 
import scipy.linalg  
# Вариант 1 решения
A = np.array([[1,2,-1],[8,-5,2]])
B = np.array([1,12])
#np.linalg.lstsq(A, B)
np.linalg.pinv(A)
np.dot(np.linalg.pinv(A),B)
```

```python
def Q(x, y, z):
  return (x**2 + y**2 + z**2)
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-50, 50, 1)
Z = np.arange(-50, 50, 1)
X, Z = np.meshgrid(X, Z)
ax.plot_surface(X, Z, Q((29+Z)/18, (0.5+0.5*Z - 29/36-Z/36), Z))
show()
```
