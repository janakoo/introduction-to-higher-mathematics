```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import math
import matplotlib 
import pylab
def fun_cos(k, a, b):
    x = np.linspace(-4*np.pi, 4*np.pi, 201)
    y = k*np.cos(x+a)+b
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.show()
fun_cos(5, 0.5*np.pi, 2)
fun_cos(1.5, np.pi, 0.5)
```
